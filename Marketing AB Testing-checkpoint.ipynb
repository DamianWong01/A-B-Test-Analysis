```python
# Null Hypothesis (H_0) - No significant difference in conversion rate between the ad and psa groups.
# Alternative Hypothesis (H_1) - There is a significant difference in conversion rate between the ad and psa groups.
```

```python
# Chi2 Test
control_data = df[df['test group'] == 'psa']
treatment_data = df[df['test group'] == 'ad']

control_conversions = control_data['converted'].sum()
treatment_conversions = treatment_data['converted'].sum()

control_total = len(control_data)
treatment_total = len(treatment_data)

control_non_conversions = control_total - control_conversions
treatment_non_conversions = treatment_total - treatment_conversions

# Contingency table
contingency_table = np.array([
    [control_conversions, control_non_conversions],
    [treatment_conversions, treatment_non_conversions]
])

# Chi-square test
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Output
print("Chi-square Test results:")
print("-" * 40)
print("Contingency Table:")
print(contingency_table)
print(f"\nChi-square statistic: {chi2:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {dof}")
print("\nExpected frequencies:")
print(expected.round())

print(f"Statistically Significant: {'Yes' if p_value < 0.05 else 'No'}")
```

```python
control_conversions = control_data['converted'].sum()
control_total = len(control_data)
control_rate = control_conversions / control_total

treatment_conversions = treatment_data['converted'].sum()
treatment_total = len(treatment_data)
treatment_rate = treatment_conversions / treatment_total

relative_lift = ((treatment_rate - control_rate) / control_rate) * 100

# Binomial test
binom_test = stats.binomtest(
    k=treatment_conversions,
    n=treatment_total,
    p=control_rate,
    alternative='two-sided'
)

print("A/B Test Results:")
print("-" * 40)
print(f"Control Group (PSA):")
print(f"Conversions: {control_conversions} out of {control_total}")
print(f"Conversion Rate: {control_rate:.2%}")

print(f"\nTreatment Group (Ad):")
print(f"Conversions: {treatment_conversions} out of {treatment_total}")
print(f"Conversion Rate: {treatment_rate:.2%}")

print(f"\nAnalysis:")
print(f"Lift: {relative_lift:.2f}%")
print(f"P-value: {binom_test.pvalue:.4f}")
print(f"Statistically Significant: {'Yes' if binom_test.pvalue < 0.05 else 'No'}")
```

```python
# Function to calculate confidence intervals
def get_confidence_interval(successes, total, confidence=0.95):
    z = stats.norm.ppf((1 + confidence) / 2)
    p = successes / total
    se = np.sqrt(p * (1 - p) / total)
    return z * se

# Calculating rates and confidence intervals
control_rate = control_conversions / control_total
treatment_rate = treatment_conversions / treatment_total

control_ci = get_confidence_interval(control_conversions, control_total)
treatment_ci = get_confidence_interval(treatment_conversions, treatment_total)

# Figure 1: Conversion Rates with Confidence Intervals
fig1, ax1 = plt.subplots(figsize=(8, 6))

groups = ['Control\n(PSA)', 'Treatment\n(Ad)']
rates = [control_rate, treatment_rate]
cis = [control_ci, treatment_ci]

bars = ax1.bar(groups, rates, yerr = cis, capsize = 5, color = ['lightgrey', '#2ecc71'])
ax1.set_title('Conversion Rates with 95% Confidence Intervals')
ax1.set_ylabel('Conversion Rate')

for i, bar in enumerate(bars):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height - 0.01,
             f'{height:.2%}',
             ha = 'center', va = 'bottom')

plt.tight_layout()
plt.show()
```

```python
# Figure 2: Null Distribution and Observed Difference
fig2, ax2 = plt.subplots(figsize=(8, 6))

n_simulations = 10000
null_differences = []

pooled_rate = (control_conversions + treatment_conversions) / (control_total + treatment_total)

# Null hypothesis
for _ in range(n_simulations):
    sim_treatment = np.random.binomial(treatment_total, pooled_rate) / treatment_total
    sim_control = np.random.binomial(control_total, pooled_rate) / control_total
    null_differences.append(sim_treatment - sim_control)

# Null distribution
sns.histplot(null_differences, ax = ax2, stat = 'density')
observed_diff = treatment_rate - control_rate

# Observed difference line
ax2.axvline(x = observed_diff, color = 'green', linestyle = '-', label = f'Observed\ndifference: {observed_diff:.3%}')
ax2.axvline(x = 0, color = 'red', linestyle = '--', label = 'Null hypothesis\n(no difference)')

ax2.set_title('Sampling Distribution Under Null Hypothesis')
ax2.set_xlabel('Difference in Conversion Rates')
ax2.set_ylabel('Density')
ax2.legend()

plt.tight_layout()
plt.show()
```
